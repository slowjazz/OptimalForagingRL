```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# State 0 == Mimic prey, State 1 == Model prey
alpha = 0.3 # Learning rate
gamma = 0.5 # Discount Factor
d0 = 3 # Constant for toxicity rejection function
epsilon = 10**-15 # Very small constant to check if learning is done
```

```python
def dFunc(t):
    """Get d probability of rejecting prey"""
    return 1-(1/(1+d0*t))

def rFunc(action, p, sampling, t):    
    """Produce the correct reward function given the parameters:
        action: action val
        p: (const) mimic pop
        sampling: 0 if no taste-sampling
        d: (const) toxic function (0) if no sampling
        t: (const) toxicity
        """
    if(action): 
        if sampling:
            d = dFunc(t)
            scale = 1/(1-(1-p)*d)
            return 2*p*scale +(1-t*t)*(1-p)*(1-d)*scale
        else:
            return 2*p+(1-t*t)*(1-p)
    else: return 1
    
def pi(Q, s, a):
    """Soft-max policy function that gives probability of taking given action at given state"""
    target = Q[s,a] 
    return np.exp(target)/sum([np.exp(Q[s,a]) for a in [0, 1]])

def pickAction(Q, s):
    """Pick an action using soft-max policy"""
    prob0 = pi(Q, 0 , 0)    
    prob1 = 1-prob0 # There are only 2 actions
    return np.random.choice(2, 1, p = [prob0, prob1] )[0]

def runQLearning(action, pop, toxicity, sampling):
    """Run Q learning for a number of iterations given relevant params
        action: starting action
        """
    R = np.array([[rFunc(0,pop,sampling, toxicity), rFunc(1,pop,sampling, toxicity)]])
    Q = np.zeros_like(R)
    A = [0,1]
    target = 800 # arbitrarily large error 
    
    while(abs(target - Q[0, action]) > epsilon):
        action = pickAction(Q, 0)
        Q_ = [Q[0,a] for a in A]
        maxQ = max(Q_)
        target = R[0, action] + gamma * maxQ # Reward of subsequent action plus discounted future reward
        TDError = alpha*(target - Q[0, action])
               
        Q[0, action] = Q[0, action] + TDError
               
    return Q
```

```python
def calcPi(T_range, p_range, sampling):
    
    X, Y = np.meshgrid(T_range, p_range)
    Z = np.zeros_like(X)
    
    interval = len(T_range)/10 # Use to print info about status of training
    count = 0
    
    for tIndex, t in enumerate(T_range):
        if(tIndex % interval <1): 
            print (str(count)+"0% Done")
            count+=1
        for pIndex, p in enumerate(p_range):
            Z[pIndex, tIndex] = pi(runQLearning(0, p, t, sampling),0,1)
    return Z

# def calcPi(T, P, sampling):
#     """Calculate the prob of attacking (Pi soft-max func) for a toxicity/pop instance and type of sampling"""
#     return pi(runQLearning(0, T, P, sampling),0,1)
```

```python
T_range = np.arange(0, 5, 0.04)
p_range = np.arange(0,1, 0.04)
X, Y = np.meshgrid(T_range, p_range)
#Z = np.vectorize(calcPi)(X, Y, 0)
Z = calcPi(T_range,p_range,0)
```

```python
plt.figure(figsize=(8,8))
pi_plot = plt.contour(X, Y, Z, (0.1, 0.3, 0.5, 0.6, 0.7), colors='black')
pi_plotFill = plt.contourf(X, Y, Z, (0, 0.5),alpha=0.2)
plt.clabel(pi_plot, inline=True, fontsize= 10)
plt.xlabel("t", fontsize = 20)
plt.ylabel("p", fontsize = 20)
plt.title("Reproduction of Figure 2");
```

```python
def piAttackToAvgReward(T, P, Pi, sampling):
    """Convert an instance of the attack probability calculation into average reward"""
    return Pi* rFunc(1, P, sampling, T) + (1-Pi)* rFunc(0, P, sampling, T)
```

```python
plt.figure(figsize=(8,8))
R = np.vectorize(piAttackToAvgReward)(X, Y, Z, 0)
R_plot_nosampling = plt.contour(X, R, Y, (0.001, 0.1, 0.25, 0.5, 0.8, 0.90), colors='black')
plt.clabel(R_plot_nosampling, inline=True, fontsize= 10, colors='r')
R_plot_nosampling.ax.fill_between([0,5],0,1, alpha = 0.1)
plt.ylim(0.6, 1.9)
plt.xlabel("t", fontsize = 20)
plt.ylabel("R", fontsize = 20)
plt.title("Reproduction of Figure 4");
```
